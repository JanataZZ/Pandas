```python
def new_price(row):
    row.price = row.price * 1.1
    return row

reviews.apply(new_price, axis='columns')
```

```python
def star_rating(row):
    if row.points >= 95:
        row['Star Rating'] = '3 star'
    if row.points >=85:
        row['Star Rating'] = '2 star'
    if row.country == 'Spain':
        row['Star Rating'] = '3 star'
    return row    
   
reviews.apply(star_rating, axis = 'columns')    

#reviews.insert(4,'Star Rating', star_rating)    
```

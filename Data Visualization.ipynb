```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
revenue = pd.read_csv('revenue.csv',index_col='Date', parse_dates=True)
```

```python
plt.figure(figsize=(14,6))
plt.title('Revenue')
sns.lineplot(data=revenue)
```

```python
plt.figure(figsize=(14,6))
plt.title('Profit Benchmarking')
sns.lineplot(data = revenue['New York'], label = 'New York')
sns.lineplot(data = revenue['Miami'], label='Miami')
```

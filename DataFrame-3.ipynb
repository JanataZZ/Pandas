```python
def convert_to_strings_and_add_millions(number):
    return str(number) + ' millions'

bond['Budget']=bond['Budget'].apply(convert_to_strings_and_add_millions)
bond['Box_Office']=bond['Box_Office'].apply(convert_to_strings_and_add_millions)
bond['Bond_Actor_Salary']=bond['Bond_Actor_Salary'].apply(convert_to_strings_and_add_millions)
```

```python
def good_movie(row):
    actor=row[2]
    budget=row[5]
    if actor == 'Sean Connery':
        return 'Best Movie'
    elif actor == 'Pierce Brosnan' and budget > 50:
        return 'Enjoyable'
    else:
        return '1 time movie'
    
bond.apply(good_movie, axis=1)    
    
```
